```python
# Add each prefix and suffix to remove to a list.
prefixes_suffixes = ["Dr. ", "Mr. ","Ms. ", "Mrs. ", "Miss ", " MD", " DDS", " DVM", " PhD"]

#replace the prefix or suffix with empty space in the student name
for word in prefixes_suffixes:
    student_data_df["student_name"] = student_data_df["student_name"].str.replace(word, "")

student_data_df.head(10)
```

```python
#merge both school_data and student_data into one dataframe using common column school_name
school_data_complete_df = pd.merge(student_data_df, school_data_df, on=["school_name", "school_name"])
school_data_complete_df
```

```python
#get the total number of students
student_count = school_data_complete_df["Student ID"].count()
student_count
```

```python
#get the total number of schools
school_count = school_data_df["School ID"].count()
school_count
```

```python
#get the unique school ids from dataframe
school_count_unique = school_data_complete_df["School ID"].unique()
len(school_count_unique)
```

```python
#calculate the total budget allocated for all schools
total_budget = school_data_df["budget"].sum()
total_budget
```

```python
# average reading score for all school
average_reading_score = school_data_complete_df["reading_score"].mean()
average_reading_score
```

```python
# average math score for all school
average_math_score = school_data_complete_df["math_score"].mean()
average_math_score
```

```python
#get passing math & reading count based on filtering on scores greater than 70
passing_math = school_data_complete_df[school_data_complete_df["math_score"] >= 70]
passing_reading = school_data_complete_df[school_data_complete_df["reading_score"] >= 70]
```

```python
passing_math_count = passing_math.student_name.count()
passing_math_count

```

```python
passing_reading_count = passing_reading.student_name.count()
passing_reading_count
```

```python
total_student_count = school_data_complete_df.student_name.count()
total_student_count
```

```python
passing_math_percentage = passing_math_count/total_student_count *100
passing_math_percentage
```

```python
passing_reading_percentage = passing_reading_count/total_student_count *100
passing_reading_percentage
```

```python
passing_math_reading = school_data_complete_df[(school_data_complete_df["math_score"] >= 70) & (school_data_complete_df["reading_score"] >= 70)]
passing_math_reading_count = passing_math_reading.student_name.count()
passing_math_reading_count
```

```python
passing_math_reading_percentage = passing_math_reading_count/total_student_count *100
passing_math_reading_percentage
```

```python
district_summary_df = pd.DataFrame(
    [{"Total Schools": school_count,
     "Total Students": total_student_count,
     "Total Budget": total_budget,
     "Average Math Score": "{:0.2f}".format(average_math_score),
     "Average Reading Score": "{:0.2f}".format(average_reading_score),
     "% Passing Math": "{:.0f}".format(passing_math_percentage),
     "% Passing Reading": "{:.0f}".format(passing_reading_percentage),
     "% Overall passing": "{:.0f}".format(passing_math_reading_percentage)}])
district_summary_df
```

```python
district_summary_df["Total Students"] = district_summary_df["Total Students"].map("{:,}".format)
district_summary_df
```

```python
district_summary_df["Total Budget"] = district_summary_df["Total Budget"].map("${:,.2f}".format)

district_summary_df["Total Budget"]
```

```python
per_school_types = school_data_df.set_index(["school_name"])["type"]
per_school_types
```

```python
df = pd.DataFrame(per_school_types)
df
```

```python
total_students_groupby = school_data_complete_df.groupby("school_name")["student_name"].count()
total_students_groupby
```

```python
total_students_valuecount = school_data_complete_df["school_name"].value_counts()
total_students_valuecount
```

```python
df["Total Students"] = total_students_groupby
df
```

```python
per_school_total_budget = school_data_df.set_index(["school_name"])["budget"]
per_school_total_budget
```

```python
df["Total School Budget"] = per_school_total_budget.map("${:,.2f}".format)
df
```

```python
per_student_budget = per_school_total_budget/total_students_valuecount
per_student_budget
```

```python
df["Per Student Budget"] = per_student_budget.map("${:,.2f}".format)
df
```

```python
average_math_score_schoolwise = school_data_complete_df.groupby("school_name")["math_score"].mean()
average_reading_score_schoolwise = school_data_complete_df.groupby("school_name")["reading_score"].mean()
```

```python
df["Average Math Score"] = average_math_score_schoolwise
df["Average Reading Score"] = average_reading_score_schoolwise
df
```

```python
#calculate count of students by school passing math
per_school_passing_math = passing_math.groupby(["school_name"])["student_name"].count()
#calculate count of students by school passing reading
per_school_passing_reading = passing_reading.groupby(["school_name"])["student_name"].count()
```

```python
#calculate passing math percentage using total students per school
per_school_passing_math_percentage = per_school_passing_math / total_students_groupby * 100
#calculate passing reading percentage using total students per school
per_school_passing_reading_percentage = per_school_passing_reading / total_students_groupby * 100

df["% Passing Math"] = per_school_passing_math_percentage
df ["% Passing Reading"] = per_school_passing_reading_percentage

df
```

```python
#calculate overall percentage passing both math and reading
#passing_math_reading
overall_passing_count = passing_math_reading.groupby(["school_name"]).count()["student_name"]
overall_passing_count
per_school_overall_passing_percentage = overall_passing_count/total_students_groupby * 100

df["% Overall Passing"] = per_school_overall_passing_percentage
df
```

```python
per_school_summary_df = pd.DataFrame({
             "School Type": per_school_types,
             "Total Students": total_students_groupby,
             "Per School Budget": per_school_total_budget.map("${:,.2f}".format),
             "Per Student Budget": per_student_budget.map("${:,.2f}".format),
             "Average Math Score": average_math_score_schoolwise,
           "Average Reading Score": average_reading_score_schoolwise,
           "% Passing Math": per_school_passing_math_percentage,
           "% Passing Reading": per_school_passing_reading_percentage,
           "% Overall Passing": per_school_overall_passing_percentage})
per_school_summary_df
```

```python
# Reorder the columns in the order you want them to appear.
new_column_order = ["School Type", "Total Students", "Total School Budget", "Per School Budget", "Average Math Score", "Average Reading Score", "% Passing Math", "% Passing Reading", "% Overall Passing"]

# Assign district summary df the new column order.
per_school_summary_df = per_school_summary_df[new_column_order]

per_school_summary_df.head()
```
